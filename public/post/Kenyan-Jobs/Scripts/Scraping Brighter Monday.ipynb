```python
import requests
import bs4
import pandas as pd
import numpy as np
```

```python
# single post
def scrape_single_post(posting_url):
    posting = requests.get(posting_url)
    posting = bs4.BeautifulSoup(posting.text, "lxml")
    try:
        title = posting.select(".job-header__title")[0].getText()
    except Exception as exc:
        title = "NA"
    try:
        poster = posting.select("h2")[0].getText()
    except Exception as exc:
        poster = "NA"
    try:
        field = posting.select("h2")[1].getText()
    except Exception as exc:
        field = "NA"
    try:
        location = posting.select(".job-header__location")[0].getText()
    except Exception as exc:
        location = "NA"
    try:
        job_type = posting.select(".job-header__work-type")[0].getText()
    except Exception as exc:
        job_type = "NA"
    try:
        industry = posting.select(".job-header__industry")[0].getText()
    except Exception as exc:
        industry = "NA"
    try:
        salary = posting.select(".job-header__salary")[0].getText()
    except Exception as exc:
        salary = "NA"
    try:
        posting_age = posting.select(".job-header__time-stamp")[0].getText()
    except Exception as exc:
        posting_age = "NA"
    try:
        summary = posting.select("p")[1].getText()
    except Exception as exc:
        summary = "NA"
    try:
        minimum_qualification = posting.select("ul li")[20].getText()
    except Exception as exc:
        minimum_qualification = "NA"
    try:
        experience_level = posting.select("ul li")[21].getText()
    except Exception as exc:
        experience_level = "NA"
    try:
        experience_length = posting.select("ul li")[22].getText()
    except Exception as exc:
        experience_length = "NA"
    try:
        qualification_1 = posting.select("ul li")[23].getText()
    except Exception as exc:
        qualification_1 = "NA"
    try:
         qualification_2 = posting.select("ul li")[24].getText()
    except Exception as exc:
         qualification_2 = "NA"
    try:
         qualification_3 = posting.select("ul li")[24].getText()
    except Exception as exc:
         qualification_3 = "NA"
    try:
         qualification_4 = posting.select("ul li")[25].getText()
    except Exception as exc:
         qualification_4 = "NA"
    try:
        qualification_5 = posting.select("ul li")[26].getText()
    except Exception as exc:
        qualification_5 = "NA"
    try:
        job_summary = posting.select(".job__details__user-edit")[0].getText()
    except Exception as exc:
        job_summary = "NA"
    try:
        job_description = posting.select(".job__details__user-edit")[1].getText()
    except Exception as exc:
        job_description = "NA"
    posting_info = [title, poster, field, location, job_type, industry, salary, posting_age, minimum_qualification, experience_length, experience_level, qualification_1, qualification_2, qualification_3, qualification_4, qualification_5, job_summary, job_description]
    return(posting_info)
```

```python
brighter_monday = pd.DataFrame(columns = ["title", "poster", "field", "location", "job_type", "industry", "salary", "posting_age", "minimum_qualification", "experience_length", "experience_level", "qualification_1", "qualification_2", "qualification_3", "qualification_4", "qualification_5", "job_summary", "job_description"])
```

```python
for i in range(1,56):
    print(i)
    page_url = "https://www.brightermonday.co.ke/jobs?page=" + str(i)
    page_info = requests.get(page_url)
    page_info = bs4.BeautifulSoup(page_info.text, "lxml")
    page_links = page_info.select(".search-result__job-title")
    for j in range(len(page_links)):
        job_link = page_links[j]["href"]
        post = scrape_single_post(job_link)
        post_df = pd.DataFrame([[post[0], post[1], post[2], post[3], post[4], post[5], post[6], post[7], post[8], post[9], post[10], post[11], post[12], post[13], post[14], post[15], post[16], post[17]]], columns = ["title", "poster", "field", "location", "job_type", "industry", "salary", "posting_age", "minimum_qualification", "experience_length", "experience_level", "qualification_1", "qualification_2", "qualification_3", "qualification_4", "qualification_5", "job_summary", "job_description"])
        brighter_monday = brighter_monday.append(post_df)
brighter_monday.index = range(len(brighter_monday))
```

```python
brighter_monday.to_csv("brighter_monday.csv", index = False)
```
